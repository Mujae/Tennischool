```python
%cd /
from glob import glob
train = glob('/content/yolov5/Tennis-School-1/train/images/*.jpg')
valid = glob('/content/yolov5/Tennis-School-1/valid/images/*.jpg')
test = glob('/content/yolov5/Tennis-School-1/test/images/*.jpg')
li = train+valid+test
```

```python
print(len(li))
```

```python
from sklearn.model_selection import train_test_split

train_img, val_img = train_test_split(li, test_size=0.2, random_state=42)
print(len(train_img), len(val_img))
```

```python
with open('/content/yolov5/train.txt', 'w') as f:
    f.write('\n'.join(train_img)+'\n')

with open('/content/yolov5/val.txt', 'w') as f:
    f.write('\n'.join(val_img)+"\n")
```

```python
import yaml

with open('/content/yolov5/Tennis-School-1/data.yaml',  'r' ) as f:
    data = yaml.full_load(f)
print(data)
data['train'] = '/content/yolov5/train.txt'
data['val'] = '/content/yolov5/val.txt'

with open('/content/yolov5/Tennis-School-1/data.yaml', 'w') as f:
    yaml.dump(data,f)

print(data)
```
